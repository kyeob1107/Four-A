```python
import pandas as pd

# 결과를 저장할 빈 데이터프레임 생성
result_df = pd.DataFrame()

# `CustomerID`별로 그룹화하고 각 그룹에 대해 merge_asof 실행
for customer_id, group in tqdm(log_sort_df.groupby('CustomerID')):
    temp_ad_click = ad_click_df[ad_click_df['CustomerID'] == customer_id][['Timestamp', 'GoodsCode']]
    merged_group = pd.merge_asof(group, temp_ad_click, on='Timestamp', tolerance=pd.Timedelta(days=7), direction='backward', suffixes=('', '_ad'))
    result_df = pd.concat([result_df, merged_group])

# 결과 확인
display(result_df.head())
```

# 추가 새로 가공

```python
log_df = pd.read_parquet("./dataset/real_log_for_spartan_optimized.parquet")
```

```python
log_eval_df = pd.read_parquet("result_df.parquet")
```

```python
log_eval_df['ActionType'].unique()
```

```python
log_eval_df
```

```python
# null값 있어서 Int16으로 했는데 이 타입 있으면 np.select typeerror나서 null->-1로 하고 타입변환
log_eval_df['GoodsCode_ad'] = log_eval_df['GoodsCode_ad'].fillna(-1)
log_eval_df['GoodsCode_ad'] = log_eval_df['GoodsCode_ad'].astype('int16')

# 조건 리스트
conditions = [
    (log_eval_df['GoodsCode_ad']==-1) & (log_eval_df['ActionType']=='view'), # not_ad_view
    (log_eval_df['GoodsCode_ad']==-1) & (log_eval_df['ActionType']=='basket'), # not_ad_basket
    (log_eval_df['GoodsCode_ad']==-1) & (log_eval_df['ActionType']=='purchase'), # not_ad_purchase
    ((log_eval_df['GoodsCode_ad']!=-1) & (log_eval_df['ActionType']=='ad_click')), # ad_click
    ((log_eval_df['GoodsCode_ad']!=-1) & (log_eval_df['ActionType']=='purchase')
     & (log_eval_df['GoodsCode']==log_eval_df['GoodsCode_ad'])), # ad_purchase
    ((log_eval_df['GoodsCode_ad']!=-1) & (log_eval_df['ActionType']=='purchase') 
     & (log_eval_df['GoodsCode']!=log_eval_df['GoodsCode_ad'])), # ad_purchase_other
    ((log_eval_df['GoodsCode_ad']!=-1) & (log_eval_df['ActionType']=='view') 
     & (log_eval_df['GoodsCode']==log_eval_df['GoodsCode_ad'])), # ad_view
    ((log_eval_df['GoodsCode_ad']!=-1) & (log_eval_df['ActionType']=='basket') 
     & (log_eval_df['GoodsCode']==log_eval_df['GoodsCode_ad'])), # ad_basket
    ((log_eval_df['GoodsCode_ad']!=-1) & (log_eval_df['ActionType']=='view') 
     & (log_eval_df['GoodsCode']!=log_eval_df['GoodsCode_ad'])), # ad_view_other
    ((log_eval_df['GoodsCode_ad']!=-1) & (log_eval_df['ActionType']=='basket') 
     & (log_eval_df['GoodsCode']!=log_eval_df['GoodsCode_ad'])), # ad_basket_other
    
]

# 조건에 따른 값 리스트
choices = ['not_ad_view', 'not_ad_basket', 'not_ad_purchase', 
           'ad_click', 'ad_purchase', 'ad_purchase_other', 
           'ad_view', 'ad_basket', 'ad_view_other', 'ad_basket_other']

# numpy.select를 사용하여 새 컬럼에 조건부 값 할당
log_eval_df = log_eval_df.assign(eval_type=np.select(conditions, choices, default='Fail'))

# 혹시 몰라 결과 공유하기 위해서 저장
log_eval_df.to_parquet("log_eval_df.parquet")
```
