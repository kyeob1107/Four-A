# **ROAS/CVR**

```python
total_ROAS = data_fin['Price'].sum() / data_fin['CPC_ad'].sum() *100
total_ROAS
```

```python
total_CVR = data_fin[data_fin['ActionType'] == 'purchase'	].count() / data_fin[data_fin['ActionType'] == 'ad_click'	].count() *100
total_CVR
```

```python
media_ROAS= data_fin.groupby('AdID_ad').agg({'Price' : 'sum' , 'CPC_ad' : 'sum' })
media_ROAS['ROAS']  = media_ROAS['Price'] / media_ROAS['CPC_ad']*100
media_ROAS
```

```python
media_CVR= data_fin.groupby('AdID_ad')['ActionType'].value_counts()
media_CVR
```

```python
media_CVR_df = media_CVR.unstack(fill_value=0).reset_index()
media_CVR_df
```

```python
media_CVR_df['CVR'] = media_CVR_df['purchase'] / media_CVR_df['ad_click']*100
```

```python
media_CVR_df
```

# **Cluster_ROAS/CVR**

```python
data_cluster= pd.read_parquet('/content/drive/MyDrive/customer_cluster_info.parquet')
```

```python
data_cluster
```

```python
merged = data_cluster.merge(data_fin, on='CustomerID' , how = 'inner')
merged
```

```python
merged_cluster_ROAS = merged.groupby('cluster_plus').agg({'Price':'sum' , 'CPC_ad' : 'sum' })
merged_cluster_ROAS
```

```python
merged_cluster_ROAS['ROAS'] = merged_cluster_ROAS['Price'] / merged_cluster_ROAS['CPC_ad']*100
```

```python
merged_cluster_ROAS
```

```python
merged_cluster_CVR = merged.groupby('cluster_plus')['ActionType'].value_counts()
merged_cluster_CVR
```

```python
merged_cluster_CVR_df = merged_cluster_CVR.unstack(fill_value=0).reset_index()
merged_cluster_CVR_df
```

```python
merged_cluster_CVR_df['CVR'] = merged_cluster_CVR_df['purchase'] / merged_cluster_CVR_df['ad_click']*100
merged_cluster_CVR_df
```

```python
# 'date' 컬럼을 datetime 형식으로 변환
merged['Timestamp'] = pd.to_datetime(merged['Timestamp'])

# 'date' 컬럼에서 연도와 월 추출하여 새로운 컬럼 생성
merged['year_month'] = merged['Timestamp'].dt.to_period('M')
```

```python
merged
```

```python
month_ROAS = merged.groupby('year_month').agg({'Price' : 'sum' , 'CPC_ad' :'sum'})
month_ROAS
```

```python
month_ROAS['ROAS'] = month_ROAS['Price'] / month_ROAS['CPC_ad']*100
month_ROAS
```

```python
month_CVR = merged.groupby('year_month')['ActionType'].value_counts()
month_CVR
```

```python
month_CVR = month_CVR.unstack(fill_value=0).reset_index()
month_CVR
```

```python
month_CVR['ad_click'] = month_CVR['ad_click'] - month_CVR['purchase']
```

```python
month_CVR['CVR'] = month_CVR['purchase'] / month_CVR['ad_click']*100
month_CVR
```
