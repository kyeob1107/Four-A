### 판매 마진율 높은 카테고리 확인

```python
sales_fin['margin_rate'] = ((sales_fin['retail_price'] - sales_fin['cost']) / sales_fin['retail_price']) * 100
sales_fin.groupby('category')['margin_rate'].mean()
```

```python
import matplotlib.pyplot as plt

# margin_rates 계산
margin_rates = sales_fin.groupby('category')['margin_rate'].mean().reset_index()

# 시각화
plt.figure(figsize=(15, 10))
bars = plt.bar(margin_rates['category'], margin_rates['margin_rate'])
plt.xlabel('Category')
plt.ylabel('Average Margin Rate')
plt.title('Average Margin Rate by Category')
plt.xticks(rotation=90)  # 카테고리 이름들을 45도 기울임

# 데이터 값 표시
for bar, rate in zip(bars, margin_rates['margin_rate']):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{rate:.2f}', ha='center', va='bottom')

plt.show()
```

### 카테고리/제품별 매출액 (Complete 만)

```python
sales_fin['total_sales'] = sales_fin['num_of_item']*sales_fin['retail_price']

complete_sales = sales_fin[sales_fin['status_x']=='Complete']

# 카테고리 별로 판매액 합산

category_sales = complete_sales.groupby('category')['total_sales'].sum().reset_index()
category_sales
```

```python
category_product_sales = complete_sales.groupby(['category', 'product_id', 'name'])['total_sales'].sum().reset_index()

# 상위 10개 제품만 
top_10 = category_product_sales.sort_values(by='total_sales', ascending=False).head(10)
top_10
```
