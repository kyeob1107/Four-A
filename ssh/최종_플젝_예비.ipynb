```python
population_2020 = {
    'California': 40000000,
    'Chicago': 2700000,
    'New Jersey': 9300000,
    'New York': 19500000,
    'Washington DC': 700000
}
```

```python
oi_new = pd.merge(oi, ci, on = '고객ID')
oi_new['총금액'] = oi_new['수량']*oi_new['평균금액']
total_revenue_by_region = oi_new.groupby('고객지역').sum('총금액').reset_index()
total_revenue_by_region.info()
```

```python
total_revenue_by_region['인구수'] = total_revenue_by_region['고객지역'].map(population_2020)
total_revenue_by_region['수량/인구수'] = total_revenue_by_region['수량']/total_revenue_by_region['인구수']
total_revenue_by_region['총금액/인구수'] = total_revenue_by_region['총금액']/total_revenue_by_region['인구수']

total_revenue_by_region
```

```python
n_customers_by_regions = oi_new.groupby('고객지역').nunique ('고객ID').reset_index()
n_customers_by_regions.rename(columns={'고객ID': '고객수'}, inplace=True)
temp = n_customers_by_regions[['고객지역','고객수']]
```

```python
final = pd.merge(temp, total_revenue_by_region, on = '고객지역')
final['고객수/인구수'] = final['고객수']/final['인구수']
final['수량/고객수'] = final['수량']/final['고객수']
final['총금액/고객수'] = final['총금액'] / final['고객수']

final
```
