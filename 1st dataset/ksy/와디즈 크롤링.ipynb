제품 타이틀

카테고리 태그

```python
import time
import random

import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup

url = "https://www.wadiz.kr/web/wreward/comingsoon"

# Chrome WebDriver 경로 설정
driver = webdriver.Chrome()

# 웹사이트 열기
driver.get(url)

# 페이지 로드를 기다림
driver.implicitly_wait(10)

data = []
for number in range(1, 11):
    # XPath를 이용해 요소 찾기
    element = driver.find_element(By.XPATH, f'//*[@id="main-app"]/div[2]/div[7]/div[2]/div[{number}]/div/a')
    
    # 요소 클릭
    element.click()
    time.sleep(random.uniform(1,3))
    # 페이지 로드를 기다림
    driver.implicitly_wait(10)
    
    # tag_xpath = '//*[@id="main-app"]/div[2]/div/div/div[2]/div[2]/div[1]/div[1]/div[1]/div[1]/div/div/a' #태그 xpath
    # second_element = WebDriverWait(driver, 10).until(
    #     EC.presence_of_element_located((By.XPATH, tag_xpath))
    #     )
    
    # 페이지의 HTML 가져오기
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    # print(soup.prettify())
    tag = soup.select('.ProjectSearchTags_hashTag__2UMo_ span')[1].text
    title = soup.find('p', class_="BaseComingSoonInfo_title__1yW_p").text
    day = soup.select('.BaseComingSoonInfo_infoWrapper__3Uo9E span')[0].text
    people = soup.find('p', class_="ComingSoonCTA_caption__SScBB").text
    
    data.append({
        'tag': tag,
        'title': title,
        'day': day,
        'people': people,
    })
    
    # 되돌아가기
    driver.back()
    time.sleep(random.uniform(1,3))

# WebDriver 종료
driver.quit()

# 데이터 프레임 생성
df_sights = pd.DataFrame(data)

# 데이터 프레임 출력
df_sights
```

```python
# WebDriver 종료
driver.quit()
```
