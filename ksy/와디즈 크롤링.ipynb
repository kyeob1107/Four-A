# 와디즈 간단히 1

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup

url = "https://www.wadiz.kr/web/wreward/main"

# Chrome WebDriver 경로 설정
driver = webdriver.Chrome()

# 웹사이트 열기
driver.get(url)

# 페이지 로드를 기다림
driver.implicitly_wait(10)

# 페이지의 HTML 가져오기
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

# 원하는 요소 선택
data = []
for info in soup.select('.HomeVerticalCard_container__3Vjdb'):
    title = info.select("strong")[0].text
    price = info.select_one(".RewardBadgeGroup_container__34pAJ").select("span")[0].text
    left_day = info.select_one(".RewardBadgeGroup_container__34pAJ").select("span")[1].text
    # print(title)
    # print(price)
    # print(left_day)
    # print("===============================================")
    data.append({
        'Title': title,
        'Price': price,
        'left_day': left_day,
    })
    
# WebDriver 종료
driver.quit()

df_funding = pd.DataFrame(data)
df_funding
```

```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup

url = "https://www.wadiz.kr/web/wreward/comingsoon"

# Chrome WebDriver 경로 설정
driver = webdriver.Chrome()

# 웹사이트 열기
driver.get(url)

# 페이지 로드를 기다림
driver.implicitly_wait(10)

data = []
for number in range(1, 11):
    # XPath를 이용해 요소 찾기
    element = driver.find_element(By.XPATH, f'//*[@id="main-app"]/div[2]/div[7]/div[2]/div[{number}]/div/a')
    
    # 요소 클릭
    element.click()
    time.sleep(1)
    # 페이지 로드를 기다림
    driver.implicitly_wait(10)
    
    # tag_xpath = '//*[@id="main-app"]/div[2]/div/div/div[2]/div[2]/div[1]/div[1]/div[1]/div[1]/div/div/a' #태그 xpath
    # second_element = WebDriverWait(driver, 10).until(
    #     EC.presence_of_element_located((By.XPATH, tag_xpath))
    #     )
    
    # 페이지의 HTML 가져오기
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    # print(soup.prettify())
    tag = soup.select('.ProjectSearchTags_hashTag__2UMo_ span')[1].text
    title = soup.find('p', class_="BaseComingSoonInfo_title__1yW_p").text
    day = soup.select('.BaseComingSoonInfo_infoWrapper__3Uo9E span')[0].text
    people = soup.find('p', class_="ComingSoonCTA_caption__SScBB").text
    
    data.append({
        'tag': tag,
        'title': title,
        'day': day,
        'people': people,
    })
    
    # 되돌아가기
    driver.back()
    time.sleep(1)

# WebDriver 종료
driver.quit()

# 데이터 프레임 생성
df_sights = pd.DataFrame(data)

# 데이터 프레임 출력
df_sights
```
