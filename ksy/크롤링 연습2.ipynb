```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import pandas as pd

url = "https://korean.visitkorea.or.kr/curation/cr_place_list.do?hotPlaceType=Place&regionCode=35"

# Chrome WebDriver 경로 설정
driver = webdriver.Chrome()

# 웹사이트 열기
driver.get(url)

# 페이지 로드를 기다림
driver.implicitly_wait(10)

# 페이지의 HTML 가져오기
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
# print(soup.prettify())
# print(soup.find_all('li', tabindex="0"))
# for tit in soup.find_all('a', class_='tit'):    
# for area in soup.find_all('div', class_ = 'area'):
data=[]
for item in soup.find_all('li', tabindex="0"):
    tit = item.select(".tit")[0].text
    area = item.select(".area em")[0].text.strip()
    # print("타이틀: ",tit)
    # print("지역: ",area)
    # break
    data.append({
        'tit': tit,
        'area': area,
    })

# WebDriver 종료
driver.quit()

# 데이터 프레임 생성
df_sights = pd.DataFrame(data)

# 데이터 프레임 출력
df_sights
```
